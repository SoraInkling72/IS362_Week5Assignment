## Question 1: What is the northernmost airport in the United States?

To be able to get the desired data, I would have to be able to generate a list of the first five northernmost airports to compare them to one another (using '.head()'), and then choose the airport on the top row. In general, the northernmost value would be the highest latitude ('lat') value in its series. So, I had to sort the DataFrame by the latitude series with descending values, so that they would be ordered from the highest value to the lowest value. By sorting the DataFrame this way, as well as having it display the first five rows, I would read the top row and see the airport name with the highest latitude value, indicating that it's the northernmost airport in the United States.

It should be noted that for latitude, positive values would indicate that the location is north of the equator in the northern hemisphere, while a negative values would indicate the southern hemosphire, south of the equator. Note, that the equator would be given the value of 0.

## Question 2: What is the easternmost airport in the United States?

While it is simpler trying to find the northernmost airport in the United States, finding the easternmost airport might be a bit more complicated. I cannot use the same method for when finding the northernmost aiport, as longitudes can be either a positive or a negative number when marking coordinates east (positive) or west (negative) from the prime meridian. And since the values can go up to either 360 or -360, it can represent either a eastern or western coordinate. Though, usually longitiude coordinates go up to 180 or -180 to avoid complications. 

I tested it for the highest and lowest values for the 'lon' series in the DataFrame, with airports listed in the top row associated with those highest/lowest values being located in Alaska, west of the mainland United States. 

While Alaska is considered both the easternmost and westernmost part of the country, according to some sources and the coordinates used. But basing this on the world map, it's actually ***westernmost***.

The easternmost airport would be located on the easternmost part of the country, which is Maine. Meaning the easternmost airport in the DataFrame would be located in Maine, but to locate it, I would have to consider another factor regarding geography: **timezones**.

To workaround this issue, I would have to take into account the correlation of the coordinate system and the timezones. First, I had to create a variable to filter the 'airports' DataFrame based on timezones less than 0, which is reprsented through the series 'tz.' This would reorder them starting from 0 and counting through negative numbers, the easternmost part of the United States has -5 as its timezone value.

Then I would sort the variable 'easternmost' based on the longitude/'lon' series, while showing the 5 lowest values for longitude as reordered as the first 5 five rows in the DataFrame. Thus, it would give me the desired information in the first row.

```python
easternmost = airports[airports.tz <= 0]
easternmost
```

```python
easternmost.sort_values('lon').tail()
```

### Answer:

Based on the information in the DataFrame generated, the easternmost airport in the United States is **Caribou Municipal Airport** in Caribou, Maine. This is because of the correlation of the timezones and longitude, given that the airport's timezone is -5 and has a longitude of 68.017917°W (-68.017917), indicating that it is the easternmost airport in the generated DataFrame. Plus, to reiterate, the easternmost part of the United States is the state of Maine, so with the airport located in the state, it can be concluded that this is the correct answer.

Overall, the correlation between the longitude ('lat') and timezone ('tz') values helped me filter the data and sort it to receive the correct answer.

## Question 3: On February 12th, 2013, which New York area airport had the windiest weather?

Much like determining the easternmost airport of the United States, I would have to apply the same principle when finding which New York area airport had the windiest weather. First, I would have to filter the DataFrame based on New York area airports, being LaGuardia Airport (LGA) and John F. Kennedy Airport (JFK). First, I had to confirm their presence in the CSV file, so I filtered both LGA and JFK in the DataFrame. Sure enough, they were there. Each DataFrame generated having 8706 rows for both airports, indicating that there are 8706 counts of recorded weather data for each airport, for almost every hour each day of the year.

```python
weather[weather.origin == 'LGA']
```

```python
weather[weather.origin == 'JFK']
```

After confirming the presence of LGA and JFK, a new DataFrame must be generated, filtered to have both LGA and JFK present in it. This new DataFrame will be used to sort the wind speed values.

```python
wind = weather[weather.origin.isin(['LGA', 'JFK'])]
wind
```

After filtering the data for LGA and JFK, the DataFrame would have to be filtered again in order to show the weather data for February 12, 2023, for both airports.

```python
date = wind[(wind.year == 2013) & (wind.month == 2) & (wind.day == 12)]
date
```

After generating this DataFrame to filter in information for February 12, 2013, another one will be generated to sort the wind speed values in a descending manner, with the first row showing the highest wind speed value. The information in that row will also show which New York area airport had the windiest weather that day.

```python
date.sort_values('wind_speed', ascending=False)
```

### Answer:

 **LaGuardia Airport** (**LGA**) is the New York area airport to have had the windiest weather on February 12, 2023 during  7 AM. Its wind speed value was 23.01560 mph, with a wind gust value of 31.07106 mph.

Considering the temperature utilizes the imperial system measurement, Fahrenheit, it can be assumed that the wind speed also utilizes imperial units as well (mph; **m**iles **p**er **h**our).
